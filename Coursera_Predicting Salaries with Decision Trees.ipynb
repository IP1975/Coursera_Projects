## Coursera - Predicting Salaries with Decision Trees

Steps:
1. Load a dataset from file
2. Process a dataset so that it can be modelled
3. Split a dataset into training and testing sets
4. Fit a decision tree to a dataset and make predictions
5. Prune a decision tree to improve its accuracy
6. Generate a graphical representation of a decision tree

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn import tree
```

```python
attribute_names = (
    'age',
    'workclass',
    'fnlwgt',
    'education',
    'education-num',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'capital-gain',
    'capital-loss',
    'hours-per-week',
    'native-country',
    'salary-class',
)
```

### 1. Load a dataset from file

```python
df = pd.read_csv("adult.data",names = attribute_names, sep=',', engine='python')
df.head()
```

### 2. Pre-process the data by removing variables that are not necessary.

```python
excludes = (
    'fnlwgt', 
    'education-num',
    'capital-gain', 
    'capital-loss'
)
for att in excludes:
    del df[att]
df.head()
```

Boolean variables to 0s and 1s.

```python
booleans = (
    'sex', 
    'native-country', 
    'salary-class'
)
for att in booleans:
    df[att] = df[att].astype('category')
    df[att] = df[att].cat.codes
df.head()
```

 One-hot encoding of all categorical attributes.

```python
categoricals = (
    'workclass', 
    'education', 
    'marital-status', 
    'occupation', 
    'relationship', 
    'race'
)
for att in categoricals:
    df = pd.concat([df,pd.get_dummies(df[att], prefix = att)], axis = 1)
    del df[att]
```

Extract attributes (X) and class labels (y).

```python
y =df['salary-class']
```

```python
X= df.copy()
X = X.drop(columns = 'salary-class') 
```

### 3. Split a dataset into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state = 34)
print('X_train', X_train.shape)
print('X_test', X_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

### 4. Fit a decision tree to a dataset and make predictions

```python
clf = DecisionTreeClassifier(random_state= 34, criterion='entropy')
clf.fit(X_train, y_train)
```

Predict target class for the training set.

```python
y_train_pred = clf.predict(X_train)
print(y_train_pred[0:10])
train_acc = accuracy_score(y_train, y_train_pred)
print("train:",train_acc)
```

Predict target class for the testing set.

```python
y_test_pred = clf.predict(X_test)
print(y_test_pred[0:10])
test_acc = accuracy_score(y_test, y_test_pred)
print("test:",test_acc)
```

Check if dataset is balanced.

```python
df['salary-class'].value_counts()
```

### 5. Prune a decision tree to improve its accuracy

```python
sizes = [2,3,5,10,20,50,100,250,500,750,1000,2500,5000,10000,15000]
train_accs = []
test_accs = []

for s in sizes:
    clf = DecisionTreeClassifier(random_state= 34,criterion='entropy', min_samples_split= s)
    clf.fit(X_train, y_train)
    
    print('samples',s)
    y_train_pred = clf.predict(X_train)
    #print(y_train_pred[0:10])
    train_acc = accuracy_score(y_train, y_train_pred)
    print("train:",train_acc)
    
    y_test_pred = clf.predict(X_test)
    #print(y_test_pred[0:10])
    test_acc = accuracy_score(y_test, y_test_pred)
    print("test:",test_acc)
    
    train_accs.append(train_acc)
    test_accs.append(test_acc)
    
plt.plot(sizes,train_accs)
plt.plot(sizes,test_accs)
plt.show()
```

### 6. Generate a graphical representation of a decision tree

```python
clf = DecisionTreeClassifier(min_samples_split= 5000,criterion= 'entropy', random_state= 34)
clf.fit(X_train, y_train)

fig= plt.figure(figsize=(40,40))
p = tree.plot_tree(clf,filled = True,feature_names= X_train.columns.to_list(),class_names=("low","high")
                  )
fig.savefig("tree.png")
```
